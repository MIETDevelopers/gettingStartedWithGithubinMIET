```python
# dropping the least important fatures
Data = Data.drop(labels=['id','Unnamed: 32'],axis=1)
```

```python
# As the diagnosis is the target i.e. the dependent feature and 
# it is in the categorical form we have to assign class numbers to them

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
Data['Target'] = labelencoder.fit_transform(Data['diagnosis'])
```

```python
Data.head()
```

```python
Data.tail()
```

```python
Data = Data.drop(labels=['diagnosis'],axis=1)
```

```python
Data.head()
```

Separating the Features and Target Variables


```python
Data_x = Data.drop(labels=['Target'],axis=1)#collection of all the Independent Features
Data_y = Data['Target']#collecton of all the target/dependent Variables
```

```python
#Separating the Test and train data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(Data_x,Data_y,test_size=0.25,random_state=42)
```

Using Accuracy score to Predict the Accuracy of Each model

```python
from sklearn.metrics import accuracy_score
```

Using the KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
```

```python
#training the Model
KNN.fit(train_x,train_y)
```

```python
#Predicting for New Values
Predict1 = KNN.predict(test_x)
```

```python
#Checking the Accuracy of the Model
Acc1 = accuracy_score(test_y,Predict1)*100
print("Accuracy of the KNN Model : ",Acc1,"%")
```

Using the Support Vector Machine

```python
from sklearn.svm import SVC
SVM = SVC()
```

```python
#training the SVM classifier
SVM.fit(train_x,train_y)
```

```python
#predicting using the Test Data
Predict2 = SVM.predict(test_x)
```

```python
#Checking the Accuracy of the Model
Acc2 = accuracy_score(test_y,Predict2)*100
print("Accuracy of the KNN Model : ",Acc2,"%")
```

Using the Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
```

```python
#training the model
DTC.fit(train_x, train_y)
```

```python
Predict3 = DTC.predict(test_x)
```

```python
#checking the accurac of the model
ACC3 = accuracy_score(test_y, Predict3)*100
print("Accuracy of the KNN model:", ACC3, "%")
```

```python
Accuracy = [Acc1, Acc2, ACC3]
plt.plot(Accuracy)
```
